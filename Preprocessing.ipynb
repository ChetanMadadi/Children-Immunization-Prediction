```python
import pandas as pd
import numpy as np

def select_k_best(X, y, k=10):
    """
    Selects the k best features from X using mutual information regression.
    """
    # Convert string values to numerical values
    for i, x_i in enumerate(X.T):
        if isinstance(x_i[0], str):
            unique_vals = np.unique(x_i)
            encoding_dict = dict(zip(unique_vals, range(len(unique_vals))))
            X[:, i] = np.vectorize(encoding_dict.get)(x_i)

    # Calculate mutual information scores
    mi_scores = np.zeros(X.shape[1])
    for i in range(X.shape[1]):
        mi_scores[i] = calc_mi(X[:, i], y)

    # Select the top k features
    top_k_idx = mi_scores.argsort()[::-1][:k]
    return X[:, top_k_idx], top_k_idx

def calc_mi(x, y):
    """
    Calculates the mutual information between x and y using the formula:
    MI(x,y) = sum( p(x,y) * log2( p(x,y) / (p(x) * p(y)) ) )
    """
    # Convert to integer values
    x_int = np.asarray(x, dtype=int)
    y_int = np.asarray(y, dtype=int)

    # Calculate the joint probability distribution
    joint_prob = np.zeros((np.unique(x_int).shape[0], np.unique(y_int).shape[0]))
    for i in range(x_int.shape[0]):
        joint_prob[x_int[i], y_int[i]] += 1
    joint_prob /= x_int.shape[0]

    # Calculate the marginal probability distributions
    x_prob = np.sum(joint_prob, axis=1)
    y_prob = np.sum(joint_prob, axis=0)

    # Calculate the mutual information
    mi = 0
    for i in range(joint_prob.shape[0]):
        for j in range(joint_prob.shape[1]):
            if joint_prob[i,j] > 0:
                mi += joint_prob[i,j] * np.log2(joint_prob[i,j] / (x_prob[i] * y_prob[j]))

    return mi

X = df
X_train_new, selected_feature_idx = select_k_best(X.astype(str).values, Y.values, k=10)
new_features = X.columns[selected_feature_idx]
print(new_features)
```
